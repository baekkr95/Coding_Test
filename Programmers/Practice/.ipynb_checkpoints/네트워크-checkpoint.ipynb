### 문제 설명

네트워크란 컴퓨터 상호 간에 정보를 교환할 수 있도록 연결된 형태를 의미합니다. 예를 들어, 컴퓨터 A와 컴퓨터 B가 직접적으로 연결되어있고, 컴퓨터 B와 컴퓨터 C가 직접적으로 연결되어 있을 때 컴퓨터 A와 컴퓨터 C도 간접적으로 연결되어 정보를 교환할 수 있습니다. 따라서 컴퓨터 A, B, C는 모두 같은 네트워크 상에 있다고 할 수 있습니다.

컴퓨터의 개수 n, 연결에 대한 정보가 담긴 2차원 배열 computers가 매개변수로 주어질 때, 네트워크의 개수를 return 하도록 solution 함수를 작성하시오.

### 깊이, 너비 우선 탐색

- 네트워크의 총 개수를 파악하는 문제
- 방문을 체크하는 visit 리스트 생성, 0으로 초기화
- 방문을 안 한 첫번째 컴퓨터부터 dfs 실행
- 자신과 인접한(연결된) 컴퓨터를 찾고 visit를 안한 상태이면 계속해서 탐색을 진행
- 탐색이 끝나면 vitis=0인 다른 컴퓨터에 대해서 dfs실행

```python
n=3
computers=[[1, 0, 0], [0, 1, 0], [0, 0, 1]]
```

### 다른 사람 풀이

```python
def solution(n, computers):
    answer = 0    # 네트워크 개수를 셀 변수
    visit = [0 for _ in range(n)]    # 노드를 방문했는지 체크하는 visit 리스트, 0으로 전부 초기화
    
    
    def dfs(start):  # 파라미터 값으로 visit를 안 한 컴퓨터 i가 넘어옴. (i = 0 ~ (n-1), n개)
        st = [start]
        print(st)
        
        while st:
            com_num = st.pop()    # com_num에 i가 들어감 (방문 안 한 컴퓨터 i)
            if visit[com_num] == 0:    # visit[i]의 방문 체크, 근데 이미 dfs함수가 방문 안한 애들이라서 사실상 필요없는 if문
                visit[com_num] = 1    # 방문 한 것으로 바꿈
            
            # 핵심 point
            for i in range(len(computers[0])):
                if computers[com_num][i] == 1 and visit[i] == 0:    # i번째 컴퓨터와 연결되고 and 방문이 안된 컴퓨터 찾기
                    print('dfs 함수 안의 for문 실행')
                    st.append(i)    # append가 되면 다시 while를 돌기 때문에 bfs를 하는 것임.
                                    # append되면서 계속 너비 탐색을 진행하는 것임. i=0 -> i=1
                                    # append 안되면 연결이 더 이상 안된거고 dfs를 나가고 다시 새로운 네트워크를 찾음.
                    
                    
    i = 0
    while 0 in visit:    # 위에서 0으로 초기화한 visit를 전부 방문(1)할 때까지 while문 돌리기
        
        if visit[i] == 0:    # visit[i]가 0이면, 방문 안했다는 뜻이므로 dfs함수 진행
            dfs(i)           # dfs함수에서 방문을 실행
            print('dfs함수 실행')
            answer += 1
            
        i += 1    # visit[i]가 이미 방문을 했으면 i+1를 해서 다음 visit[i+1]에 대해서 방문을 체크
        
        
    return answer
```

```python
solution(n, computers)
```
