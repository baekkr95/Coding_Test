### 문제 설명
도둑이 어느 마을을 털 계획을 하고 있습니다. 이 마을의 모든 집들은 아래 그림과 같이 동그랗게 배치되어 있습니다.
각 집들은 서로 인접한 집들과 방범장치가 연결되어 있기 때문에 인접한 두 집을 털면 경보가 울립니다.

각 집에 있는 돈이 담긴 배열 money가 주어질 때, 도둑이 훔칠 수 있는 돈의 최댓값을 return 하도록 solution 함수를 작성하세요.

### 문제 풀이

- 첫번째 집을 턴다 / 첫번째 집을 안 턴다 -> 2가지로 나뉨.
- 집을 털 때마다 돈의 최댓값을 저장할 table을 생성
- n번 째 집 = max( (n-1)번까지 턴 돈, (n-2)번까지 턴 돈 + n번째 집 돈 ) -> 점화식

```python
def solution(money):
    # 새로운 배열 생성해서 최대값 계속 갱신
    # 점화식: max(i번째집+(i-2)까지 턴 돈, (i-1)까지 턴 돈)
    # 첫번째 집을 터는가? 아닌가?로 나뉜다.
    # 만약 첫번째 집을 털면 마지막집 못 턴다.
    table = [0 for _ in range(len(money))]
    table[0] = money[0]
    table[1] = 0
    for i in range(2, len(money)-1): # 마지막집 안 터니까 -1을 한다.
        table[i] = max(table[i-1], table[i-2]+money[i])
    
    
    table2 = [0 for _ in range(len(money))]
    table2[1] = money[1]
    for i in range(2, len(money)):
        table2[i] = max(table2[i-1], table2[i-2]+money[i])
        
    return max(max(table), max(table2))
```

```python
money = [1, 2, 3, 1]
solution(money)
```
